# 2. CURĂȚAREA ȘI PRELUCRAREA BAZEI DE DATE

2.1. ELIMINAREA CUVINTELOR DE OPRIRE

```python
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('romanian')
print(stopword)

from nltk.corpus import stopwords
", ".join(stopwords.words('romanian'))
STOPWORDS = set(stopwords.words('romanian'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

tweet_df["text_stop_word"] = tweet_df["comentariu"].apply(lambda text: remove_stopwords(text))
tweet_df.head()
```

```python
%matplotlib inline
pd.set_option('display.max_colwidth', 100)

tweet_df.head(20)
```

2.2. ELIMINAREA SEMNELOR DE PUNCTUAȚIE

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

tweet_df["text_stop_word"] = tweet_df["text_stop_word"].apply(lambda x: remove_punct(x))
tweet_df.head(10)
```

2.3. REALIZAREA OPERAȚIUNII DE STEMMING ASUPRA SETULUI DE DATE

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
tweet_df["text_stemmed"] = tweet_df["text_stop_word"].apply(lambda text: stem_words(text))
tweet_df.head()
```

2.4. IDENTIFICAREA ȘI ELIMINAREA CELOR MAI FRECVENTE CUVINTE DIN SETUL DE DATE

```python
from collections import Counter
cnt = Counter()
for text in tweet_df["text_stemmed"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(20)

FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

tweet_df["text__stopfreq"] = tweet_df["text_stemmed"].apply(lambda text: remove_freqwords(text))
tweet_df.head()
```

2.5. REALIZAREA OPERAȚIUNII DE LEMNATIZARE A SETULUI DE DATE

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

tweet_df["text_lemmatized"] = tweet_df["text__stopfreq"].apply(lambda text: lemmatize_words(text))
tweet_df.head()
```

2.6. CODIFICAREA VARIABILEI CATEGORIALE

```python
tweet_df.categorie.replace('neutru' , 0 ,inplace = True)
tweet_df.categorie.replace('pozitiv' , 1 ,inplace = True)
tweet_df.categorie.replace('negativ', -1 ,inplace = True)
tweet_df.head()
```

2.7. ELIMINAREA COLOANELOR NEUTILIZATE
